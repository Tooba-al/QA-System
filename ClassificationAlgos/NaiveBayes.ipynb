```python
df = pd.read_csv("ClassificationAlgos/Features_dev1.csv", encoding="cp1252")
df = df.loc[:, df.columns != "syntatic_divergence"]
df = df.loc[:, df.columns != "consistant_labels"]
df = df.loc[:, df.columns != "span_POS_tags"]
```

```python
categorical = [var for var in df.columns if df[var].dtype == "O"]
print(Fore.GREEN + "There are {} categorical variables.".format(len(categorical)))
print(Fore.GREEN + "The categorical variables are :\n", categorical)

df["wh_word"].replace("", np.NaN, inplace=True)
print(Fore.GREEN + "Frequency distribution of values:")
print(df["wh_word"].value_counts() / np.float(len(df)))
print("Missing values in {} variables.".format(df["wh_word"].isnull().sum()))
# print("check labels :" + str(train_df[var].unique()))
print(
    "Cardinality : ",
    "wh_word",
    " contains ",
    len(df["wh_word"].unique()),
    " labels.",
)
```

```python
numerical = [var for var in df.columns if df[var].dtype != "O"]
print(Fore.GREEN + "There are {} numerical variables\n".format(len(numerical)))
print("The numerical variables are :", numerical)
# check missing values in numerical variables
print("Missing Values in in {} variables".format(df[numerical].isnull().sum()))
```

```python
# Declare feature vector and target variable
X = df.drop(["answer"], axis=1)
y = df["answer"]

# Split data into separate training and test set
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0
)

# check the shape of X_train and X_test
X_train.shape, X_test.shape
```

```python
# impute missing categorical variables with most frequent value

for df2 in [X_train, X_test]:
    df2["question"].fillna(X_train["question"].mode()[0], inplace=True)
    df2["span"].fillna(X_train["span"].mode()[0], inplace=True)
    df2["answer"].fillna(X_train["answer"].mode()[0], inplace=True)
    df2["wh_word"].fillna(X_train["wh_word"].mode()[0], inplace=True)
```
